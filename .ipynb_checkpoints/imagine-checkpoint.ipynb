```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load data in a Dataframe
# read the first sheet into a dataframe
df1 = pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2009-2010")

# read the second sheet into a dataframe
df2 = pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2010-2011")

# concatenate the two dataframes vertically using the `concat` function
df = pd.concat([df1, df2], ignore_index=True)
```

```python
# Check the first records of the data
df.head(10)
```

```python
# Check the last records of the data
df.tail(10)
```

```python
# Conver data types
df.convert_dtypes().dtypes
```

```python
# group the data by country and sum the quantity for each country
country_data = df.groupby("Country")["Quantity"].sum().reset_index()

# sort the data in descending order based on the quantity
country_data = country_data.sort_values("Quantity", ascending=False)

# show the most consumed country
most_consumed_country = country_data.loc[
    country_data["Quantity"].idxmax(), "Country"
]
most_consumed_quanty = country_data.loc[
    country_data["Quantity"].idxmax(), "Quantity"
]
print(
    f"The top consuming is {most_consumed_country} with {most_consumed_quanty}"
)

# create a bar plot to visualize the data
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(country_data["Country"], country_data["Quantity"])
ax.set_xlabel("Country")
ax.set_ylabel("Quantity")
ax.set_title("Country Consumption")
ax.tick_params(axis="x", labelrotation=90)
plt.show()
```

```python
# group the data by StockCode and sum the Quantity
product_sales = df.groupby(["StockCode"])["Quantity"].sum().reset_index()
result_sorted = product_sales.sort_values(
    "Quantity", ascending=False
).reset_index(drop=True)
top_products = result_sorted.head(10)
top_products
```

```python
# Merge with the original dataframe to get the description
top_products_with_des = pd.merge(
    top_products,
    df[["StockCode", "Description"]].drop_duplicates(subset=["StockCode"]),
    on="StockCode",
    how="left",
)

# Display the result
print(top_products_with_des)
```

```python
plt.bar(
    top_products_with_des["Description"], top_products_with_des["Quantity"]
)
plt.xticks(rotation=90)
plt.xlabel("Product Description")
plt.ylabel("Sales Quantity")
plt.title("Top 10 Best-Selling Products")
plt.show()
```

```python
# Add a TotalPrice column
df["TotalPrice"] = df["Quantity"] * df["Price"]

# Group the data by year and month and sum the TotalPrice column
monthly_sales = df.groupby(
    [df["InvoiceDate"].dt.year, df["InvoiceDate"].dt.month]
)["TotalPrice"].sum()

# Create a line chart of monthly sales for each year
monthly_sales.unstack(level=0).plot(kind="line", figsize=(10, 6))

# Set the chart title and axes labels
plt.title("Total Monthly Sales by Year")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Select the variables for the scatter plot
x = df["Quantity"]
y = df["Price"]

corr_coef = y.corr(x)
# Create the scatter plot
plt.scatter(x, y, alpha=0.5)

# Add labels and title
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.title(f"Correlation between Price and Quantity: {corr_coef:.2f}")

# Show the plot
plt.show()
```

```python
# Extract month from InvoiceDate column
df["Month"] = pd.to_datetime(df["InvoiceDate"]).dt.month

# Calculate correlation coefficient
corr_coef = df["Month"].corr(df["Quantity"])

# Plot scatter plot
plt.scatter(df["Month"], df["Quantity"])
plt.xlabel("Month")
plt.ylabel("Quantity")
plt.title(f"Correlation Coefficient: {corr_coef:.2f}")
plt.show()
```

```python
# Calculate the total sales by country
sales_by_country = df.groupby('Country')['Quantity'].sum().reset_index()

# Calculate the total sales by country and month
df['Month'] = df['InvoiceDate'].dt.month
sales_by_country_month = df.groupby(['Country', 'Month'])['Quantity'].sum().reset_index()

# Merge the two datasets
merged_data = pd.merge(sales_by_country_month, sales_by_country, on='Country', how='left')
merged_data.rename(columns={'Quantity_x': 'Sales_by_month', 'Quantity_y': 'Total_sales'}, inplace=True)

# Calculate the correlation coefficient between sales by month and total sales
correlation_coefficient = merged_data['Sales_by_month'].corr(merged_data['Total_sales'])

# Plot the data using a scatterplot
plt.scatter(merged_data['Total_sales'], merged_data['Sales_by_month'])
plt.xlabel('Total sales by country')
plt.ylabel('Sales by month')
plt.title(f'Correlation coefficient: {correlation_coefficient:.2f}')
plt.show()
```
